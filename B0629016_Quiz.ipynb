```python
import matplotlib.pylab as pl
x= Weijin_SawtoothSignal(freq=1000,amp=10)
x.plot()
pl.grid()
```

```python
w= x.make_wave()
w.segment(duration=.1).plot()
w.make_audio()
```

```python
s= w.make_spectrum()
s.plot(high=1000)
```
